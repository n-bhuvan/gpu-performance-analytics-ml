```python
import pandas as pd
import matplotlib.pyplot as plt

reg_results = pd.DataFrame({
    "Model":[
        "Linear",
        "Ridge",
        "Lasso",
        "Random Forest",
        "XGBoost",
        "Tuned XGBoost"
    ],
    "R2":[
        0.8775,
        0.8774,
        0.8774,
        0.9835,
        0.9880,
        0.9905
    ]
})

plt.figure(figsize=(8,5))
plt.bar(reg_results["Model"], reg_results["R2"])
plt.title("Regression Model Comparison")
plt.ylabel("R² Score")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
clf_results = pd.DataFrame({
    "Model":[
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],
    "Accuracy":[
        0.7854,
        0.9614,
        0.9871
    ]
})

plt.figure(figsize=(8,5))
plt.bar(clf_results["Model"], clf_results["Accuracy"])
plt.title("Classification Model Comparison")
plt.ylabel("Accuracy")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

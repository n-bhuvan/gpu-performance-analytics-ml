```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/feature_engineered_gpu_dataset.csv")

df.head()
```

```python
df.dtypes
```

```python
selected_columns = [
    'manufacturer',
    'releaseyear',
    'memsize',
    'gpuclock',
    'memclock',
    'unifiedshader',
    'tmu',
    'rop',
    'g3dmark',
    'g2dmark',
    'price',
    'gpuvalue',
    'tdp',
    'powerperformance',
    'opencl',
    'performance_per_watt',
    'performance_per_gb',
    'performance_per_shader',
    'gpu_age',
    'performance_category'
]

df = df[selected_columns]
```

```python
from sklearn.preprocessing import LabelEncoder

manufacturer_encoder = LabelEncoder()

df["manufacturer"] = manufacturer_encoder.fit_transform(
    df["manufacturer"]
)

df.head()
```

```python
X_reg = df.drop(
    columns=[
        "g3dmark",
        "performance_category"
    ]
)

y_reg = df["g3dmark"]
```

```python
performance_encoder = LabelEncoder()

df["performance_category"] = performance_encoder.fit_transform(
    df["performance_category"]
)

X_clf = df.drop("performance_category", axis=1)

y_clf = df["performance_category"]
```

```python
from sklearn.model_selection import train_test_split

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.20,
    random_state=42
)
```

```python
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf,
    y_clf,
    test_size=0.20,
    random_state=42,
    stratify=y_clf
)
```

```python
from sklearn.preprocessing import StandardScaler

reg_scaler = StandardScaler()

X_train_reg_scaled = reg_scaler.fit_transform(
    X_train_reg
)

X_test_reg_scaled = reg_scaler.transform(
    X_test_reg
)
```
